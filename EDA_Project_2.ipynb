
# <p style="background-color:#E2E3E5;font-family:newtimeroman;font-size:50px;color:#26306B;text-align:Left;border: 2px solid #000000;"><b>How Does the Funding Ecosystem changes with respect to Time?</b></p>

Zamana göre finansman ekosistemini incelememiz için tarihlere göre yatırım yapılmasının kırılımını yapmalıyız. Bunun için zaman columnunda düzenleme yapıyoruz.

```python
# Tarih columnundaki gerekli düzeltmeleri yapıyoruz.

df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="12/05.2015"] = "12/05/2015"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="13/04.2015"] = "13/04/2015"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="15/01.2015"] = "15/01/2015"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="22/01//2015"] = "22/01/2015"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="05/072018"] = "05/07/2018"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="01/07/015"] = "01/07/2015"
df["Date_dd/mm/yyyy"][df["Date_dd/mm/yyyy"]=="\\\\xc2\\\\xa010/7/2015"] = "10/07/2015"
```

```python
# Yıl bazında analiz yapmak için ayrı bir column oluşturuyoruz.

df["Date"] = df["Date_dd/mm/yyyy"] = pd.to_datetime(df["Date_dd/mm/yyyy"], format="%d/%m/%Y", errors="coerce")
```

```python
# Yapılan yatırımı tarih-ay bazında analiz etmek için yeni bir column oluşturuyoruz. Ay bazında yapılan yatırımı count yapıyoruz.

df["yearmonth"] = (pd.to_datetime(df["Date_dd/mm/yyyy"], format="%d/%m/%Y").dt.year*100)+(pd.to_datetime(df["Date_dd/mm/yyyy"],format="%d/%m/%Y").dt.month)
funding_time = df["yearmonth"].value_counts().sort_values(ascending = False)
funding_time
```

```python
fig, ax = plt.subplots(figsize=(12,6))
plt.title(label="The Funding Ecosystem Change Over Time", size= 15, color="black")

sns.barplot(x=funding_time.index, y=funding_time.values.flatten(), palette="Spectral")
plt.xticks(color="black", rotation=90, size=9);
plt.xlabel("", size=12, color="black")
ax.bar_label(ax.containers[0])
plt.show()
```

### **GRAFİK YORUMU**

Grafikte, 2015'in başından 2020'nin başına kadar olan dönemde startup ekosistemindeki fonlama değişiklikleri gösterilmektedir. X ekseni zaman dilimlerini (yıl ve ay bazında), Y ekseni ise belirli bir zaman diliminde gerçekleşen fonlama olaylarının (yatırım turu sayısı, fonlanan startup sayısı vb.) miktarını göstermektedir.


**2015 Yılı Başlangıcı:**
2015'in ilk yarısında fonlama olaylarında istikrarlı bir artış görülüyor. Örneğin, Ocak 2015'te yaklaşık 53 fonlama olayı varken, Haziran 2015'te bu sayı 86'ya yükseliyor.


**2015 ve 2016 Yılları:**
2015 ve 2016 yılları arasında fonlama olaylarında genel olarak yüksek bir artış trendi var. En yüksek nokta, Ocak 2016'daki 104 fonlama olayı ile görülüyor.    
Bu yıllar arasında startup ekosistemi oldukça hareketli ve yatırımcı ilgisi yüksek görünüyor.


**2016 Orta ve Son Çeyreği:**
2016 yılının ortalarında fonlama olaylarında bir düşüş başlıyor. Temmuz 2016'da 64 olaydan, Eylül 2016'da 72 olaya ve Kasım 2016'da 71 olaya kadar dalgalanma var.
Genel trend, bu dönemde bir miktar düşüş eğiliminde.
Hindistan'da 2016 yılında, Hindistan Merkez Bankası tarafından ani olarak ilan edilen ve likidite sıkıntısı yaratan demonetizasyon politikası uygulandı. Bu politika, 500 ve 1000 rupi değerindeki banknotların aniden iptal edilmesini içeriyordu. Bu politikanın uygulanması, Hindistan'daki tüketim harcamalarını ve ekonomik aktiviteyi geçici olarak olumsuz etkiledi.


**2017 Yılı:**
2017 yılının başında ve ortasında fonlama olaylarında bir düşüş gözleniyor. Örneğin, Ocak 2017'de 48 olaydan, Haziran 2017'de 65 olaya kadar.
Ancak yılın sonlarına doğru, Eylül 2017'de 43, Kasım 2017'de 45 fonlama olayı ile yine bir toparlanma görülüyor.


**2018 ve 2019 Yılları:**
2018 ve 2019 yıllarında belirgin bir düşüş trendi var. Ocak 2018'de 37 olaydan başlayarak, Aralık 2018'de 7 olaya kadar düşüş devam ediyor.
2019 yılında ise biraz toparlanma var ancak genel olarak 2016 ve 2017 seviyelerine geri dönülemiyor. Örneğin, Eylül 2019'da 17 olay görülüyor.


**2020 Başlangıcı:**
Ocak 2020'de fonlama olayları 7'ye kadar düşmüş, bu da 2015-2020 döneminin en düşük noktalarından biri. Özellikle 2020'deki düşüş, COVID-19 salgını ve pandemik koşulların etkisiyle yaşanmış olabilir.

```python
# Finansman türlerini Yatırım tutarlarına göre listeliyoruz.

Investment_type_usd = df.groupby("InvestmentnType")["Amount_in_USD"].sum().sort_values(ascending=False).head(10)
Investment_type_usd
```

```python

# İki grafik oluşturdum.
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 6))

# Birinci grafik
sns.barplot(x=Investment_type.index, y=Investment_type.values.flatten(), palette="tab10", ax=ax[0])
ax[0].set_title("Most Populer Funding Types", size=15,)
ax[0].set_xticklabels(ax[0].get_xticklabels(), ha="center",  rotation=45, size=10)
ax[0].set_xlabel("", size=12, color="black")
for p in ax[0].patches:
    ax[0].annotate(str(int(p.get_height())), (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha="center", va="center", fontsize=9, color="black", xytext=(0, 5),
                   textcoords="offset points")


# İkinci grafik
sns.barplot(x=Investment_type_usd.values.flatten(), y=Investment_type_usd.index, palette="tab10", ax=ax[1])
ax[1].set_title("Most Populer Funding Types by Amount USD", size=15)
ax[1].set_xticklabels(ax[1].get_xticklabels(),ha="right", size=10)
ax[1].set_ylabel("", size=12, color="black")


plt.tight_layout()
plt.show()
```
